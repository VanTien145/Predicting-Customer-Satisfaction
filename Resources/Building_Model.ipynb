# PROCESS AND TRANFORM

## Split Dataset


```python
# modelling
from sklearn.model_selection import train_test_split
```

```python
# Save df as df_copy to train model in second way
df_copy = df.copy()
```

```python
df.replace(['dissatisfied', 'satisfied'], [0,1], inplace=True)
df.replace(['dissatisfied', 'satisfied'], [0,1], inplace=True)
df.replace(['Eco', 'Eco Plus', 'Business'], [0,1,2], inplace=True)
df.replace(['Eco', 'Eco Plus', 'Business'], [0,1,2], inplace=True)
```

```python
X, y = df.drop(columns='satisfaction'), df['satisfaction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape
```

## Encode and Scale Data

```python
# data processing
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
```

```python
one_hot_feats = sorted(['gender', 'customer_type', 'type_of_travel'])
numeric_feats = sorted(['age', 'flight_distance', 'seat_comfort',
       'departure_or_arrival_time_convenient', 'food_and_drink',
       'gate_location', 'inflight_wifi_service', 'inflight_entertainment',
       'online_support', 'ease_of_online_booking', 'on_board_service',
       'leg_room_service', 'baggage_handling', 'checkin_service',
       'cleanliness', 'online_boarding', 'departure_delay_in_minutes', 'arrival_delay_in_minutes', 'class'])

# define transformation
one_hot_trans = OneHotEncoder()
numeric_trans = MinMaxScaler()

data_trans = ColumnTransformer(transformers=[
    ('one_hot', one_hot_trans, one_hot_feats),
    ('numeric', numeric_trans, numeric_feats),
])
```

```python
temp = data_trans.fit_transform(X_train)
df_train_trans = pd.DataFrame(temp, index=X_train.index, columns=data_trans.get_feature_names_out())
```

```python
temp = data_trans.transform(X_test)
df_test_trans = pd.DataFrame(temp, index=X_test.index, columns=data_trans.get_feature_names_out())
```

```python
# Check imbalance
y_train.value_counts()
```

# TRAIN & EVALUATE MODEL

## Logistic Regression


```python
# modelling
from sklearn.linear_model import LogisticRegression

# metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model_log = LogisticRegression()
model_log.fit(df_train_trans, y_train)
predict = model_log.predict(df_test_trans)


# Result
print(classification_report(y_test, predict))
```

## XGBoost

```python
from sklearn.ensemble import (RandomForestClassifier, AdaBoostClassifier,
                              GradientBoostingClassifier)
import xgboost as xgb
```

```python
model_xgb = xgb.XGBClassifier()
model_xgb.fit(df_train_trans, y_train)
predict = model_xgb.predict(df_test_trans)


# Result
print(classification_report(y_test, predict))
```

## AdaBoost

```python
from sklearn.tree import (DecisionTreeClassifier, ExtraTreeClassifier)
```

```python
# define
et_clf = ExtraTreeClassifier(
    criterion='gini',
    max_depth=4,
    min_samples_split=2,
    min_samples_leaf=1,
    max_leaf_nodes=None,
    max_features=None,
)
ada_clf = AdaBoostClassifier(
    estimator=et_clf,
    n_estimators=50,
    learning_rate=1.0,
    random_state=33,
)
# train
ada_clf.fit(df_train_trans, y_train)
# eval
y_pred = ada_clf.predict(df_test_trans)
print(classification_report(y_test, y_pred))
```

## gradient boost tree

```python
# define
gbt_clf = GradientBoostingClassifier(
    n_estimators=50,
    learning_rate=1.0,
    random_state=33,
)
# train
gbt_clf.fit(df_train_trans, y_train)
# eval
y_pred = gbt_clf.predict(df_test_trans)
print(classification_report(y_test, y_pred))
```
